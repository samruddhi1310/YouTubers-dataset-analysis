#Q.1)What is the distribution of rank among youtubers?

```python
num_youtubers=1000

alpha=2.0
xmin=1
Followers=xmin*(np.random.pareto(alpha,num_youtubers))

sort_subs=np.sort(Followers)[::-1]

for rank,subs_count in enumerate(sort_subs,start=1):
  print(f"Rank {rank}: {subs_count:.2f} Followers")
```

#Conclusion

This code basically generates a power law distribution of subscribers counts among YouTubers and prints the distribution of rank along their corresponding subscribers count

#Q.8)Is there any geographical patterns in terms of where Youtubers with high subscribers or engagement are located?

```python
country=yt.groupby('Country').agg({'Followers':'mean','Likes':'mean','Comments':'mean'})

fig,axes=plt.subplots(nrows=3 ,figsize=(30,15))

# Plotting mean subscribers
axes[0].bar(country.index,country['Followers'])
axes[0].set_title('Mean Subscribers by Country')
axes[0].set_xlabel('Country')
axes[0].set_ylabel('Mean Subscribers')

# Plotting mean likes
axes[1].bar(country.index,country['Likes'])
axes[1].set_title('Mean Likes by Country')
axes[1].set_xlabel('Country')
axes[1].set_ylabel('Mean Likes')

# Plotting mean comments
axes[2].bar(country.index,country['Comments'])
axes[2].set_title('Mean Comments by Country')
axes[2].set_xlabel('Country')
axes[2].set_ylabel('Mean Comments')

plt.tight_layout()
plt.show()

```

#Conclusion

This code basically groups the data by country,calculates the mean number of subscribers,likes and comments for each country,and then plots these metrics on the separate charts.and by examining these visualization we can identify any geographical patterns in terms of whwre Youtubers with high subscribers are located

#Data Prediction

For the prediction of dataset of "Youtubers" containing row:
Rank, Username,	Categories,	Suscribers,	Country,	Visits,	Likes,	Comments, and	Links we will predict the data

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
yt=pd.read_csv("/content/youtubers_df[1].csv")
```

```python
yt
```

```python
print(yt.dtypes)
```

#Predictions

```python
x=yt[["Suscribers","Country","Comments","Visits","Links","Categories"]]
y=yt["Likes"]

x=pd.get_dummies(x)
x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(x_train,y_train)

y_pred=model.predict(x_test)

mse=mean_squared_error(y_test,y_pred)
print("Mean Squared Error :",mse)
```

#Model training

```python
x=yt[['Suscribers','Visits','Likes']] #Stored in new DataFrame named x
y=yt['Comments'] #Comments column stored in new DataFrame named y

x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=0.2, random_state=42)

model=LinearRegression() #Creates LinearRegression model object

model.fit(x_train, y_train) #trains model using fit

#To evaluate the model
train_score=model.score(x_train, y_train)
test_score=model.score(x_test,y_test)

print("Training score:",train_score)
print("Testing score:",test_score)  #Evaluates the model's performance on both the training and testing sets
```

```python
print(y_train.head())
y_train = y_train[y_train != '59729\uf0d7']
y_test = y_test[y_test != '59729\uf0d7']
y_train = pd.to_numeric(y_train, errors='coerce')
y_test = pd.to_numeric(y_test, errors='coerce')
model.fit(x_train, y_train)
train_score = model.score(x_train, y_train)
test_score = model.score(x_test, y_test)

print("Training score:", train_score)
print("Testing score:", test_score)
```
